# 1. Introduction

This project analyses the relationship between firm size, financial leverage, and profitability using financial data from Chinese listed companies.

The objective of this analysis is to explore how firm size and leverage influence profitability, and to identify potential trade-offs between financial return and risk.

# 2. Data Description

The dataset is obtained from the CSMAR database and contains financial information on Chinese listed companies.

For analytical purposes, firms are grouped into four size categories: Low, Medium-Low, Medium-High, and High. This grouping simplifies comparison across firms of different scales.

The key variables used in this analysis include:

- Net Profit: Measures the total earnings of a firm
- ROA (Return on Assets): Indicates how efficiently a firm uses its assets to generate profit
- ROE (Return on Equity): Reflects the return generated for shareholders
- Leverage: Represents the level of financial risk based on debt usage
- Profit Margin: Measures operational efficiency and profitability

Although the original dataset is large, aggregating firms into size groups reduces the number of observations, which may affect the robustness of the analysis.

```python
# Inspect the column structure of the CSMAR financial table 
financial_desc = db.describe_table(library="csmar", table="wrds_csmar_financial_master")
financial_desc
```

```python
# Select only the 3 most useful columns
financial_desc_small = financial_desc[["name", "type", "comment"]]

# Show the first 20 rows as a preview.
financial_desc_small.head(20)
```

```python
# Define the list of variable IDs we plan to use.
ids_today = [
    "a001000000",   # Total Assets
    "a002000000",   # Total Liabilities
    "a003000000",   # Total Shareholders' Equity
    "b001100000",   # Total Operating Revenue
    "b002000000"    # Net Profit
]
financial_desc_selected = financial_desc_small[financial_desc_small["name"].isin(ids_today)]

# Display the filtered description table.
financial_desc_selected
```

```python
# Store columns as a list so we can loop over them or join them into a SQL string.
financial_columns_list = ["stkcd", "accper", "typrep","a001000000","a002000000","a003000000", "b001100000", "b002000000"]
```

```python
# Converts the format before we put it into the query.
financial_columns = ", ".join(financial_columns_list)

# Print the result to verify it looks like a valid SQL column list.
print(financial_columns)
print(type(financial_columns))
```

```python
# --- Build the SQL query for Chinese financial data using an f-string ---
financial_query = f"""
SELECT {financial_columns}
FROM csmar.wrds_csmar_financial_master
WHERE accper >= '2020-01-01'
AND typrep = 'A'
"""
print(financial_query)
```

```python
china_financial = db.raw_sql(financial_query, date_cols=["accper"])
china_financial.head()
```

```python
# Add columns to calculate size, ROA, ROE, leverage and profit margin
# ROA: profitability
# ROE: shareholder return
# Leverage: financial risk
# profit margin: revenue
china_financial['size'] = np.log(china_financial["a001000000"])
china_financial["ROA"] = china_financial["b002000000"] / china_financial["a001000000"]
china_financial["ROE"] = china_financial["b002000000"] / china_financial["a003000000"]
china_financial['leverage'] = china_financial["a002000000"] /china_financial["a001000000"]
china_financial['profit margin'] = china_financial["b002000000"] /china_financial["b001100000"]
china_financial
```

```python
china_financial["report_year"]  = china_financial["accper"].dt.year
china_financial["report_month"] = china_financial["accper"].dt.month
china_financial["report_day"]   = china_financial["accper"].dt.day
china_financial
```

```python
# Keep only year-end (December) reports
china_financial_annual = china_financial[china_financial['report_month'] == 12].copy()

china_financial_annual
```

```python
# The result shows us exactly what each of our chosen columns means.

selected_desc = financial_desc_small[financial_desc_small["name"].isin(financial_columns_list)]

# Display the filtered description — this is the reference for building rename_map.
selected_desc
```

This section presents visualisations to compare financial performance across different firm size groups.

In contrast, the High group exhibits a wider spread and a large number of outliers, suggesting that very large firms differ substantially even within the same category. This reflects the presence of extreme values and greater heterogeneity among large firms.

# 4.Regression analysis

```python
# Select columns to make regression analysis
selected_regression = selected_view[['Net Profit','ROA', 'ROE', 'leverage','profit margin']]
selected_regression
```

```python
import pandas as pd
import statsmodels.api as sm
df = selected_regression.copy()

df = df[['ROA', 'leverage', 'profit margin']]


for col in df.columns:
    df[col] = df[col].astype(str).str.strip()   
    df[col] = pd.to_numeric(df[col], errors='coerce')


df = df.dropna()


print(df.dtypes)


Y = df['ROA']
X = df[['leverage', 'profit margin']]

X = sm.add_constant(X)

model = sm.OLS(Y, X).fit()
print(model.summary())
```

```python
import matplotlib.pyplot as plt

beta = model.params.drop('const')

plt.figure(figsize=(6,4))
plt.bar(beta.index, beta.values)

plt.axhline(0)

plt.title('Regression Coefficients (β)')
plt.xlabel('Variables')
plt.ylabel('Coefficient Value')

plt.show()
```

The regression results examine the relationship between leverage, profit margin, and firm profitability (ROA).

The model shows a moderate explanatory power, with an R-squared value of 0.581, indicating that approximately 58.1% of the variation in ROA can be explained by the independent variables. The overall model is statistically significant, as indicated by the F-statistic (p < 0.01).

Leverage has a statistically significant negative effect on ROA (coefficient = -0.1541, p < 0.01). This suggests that firms with higher financial leverage tend to have lower profitability, likely due to increased financial risk and interest burden. This is the most important finding of the analysis.

In contrast, profit margin has a positive but very small coefficient and is only marginally significant at the 10% level (p = 0.063). This indicates that while operational efficiency may contribute to profitability, its impact is relatively weak compared to leverage.

However, diagnostic statistics reveal potential limitations. The residuals are highly skewed and exhibit extreme kurtosis, indicating the presence of outliers and a violation of the normality assumption. This may affect the reliability of the estimated coefficients.

Overall, the results highlight that financial risk plays a more significant role than profit margin in explaining firm profitability. However, the findings should be interpreted with caution due to potential data distribution issues.

# 6.Limitations

Several limitations should be considered when interpreting the results of this analysis.

First, although the original dataset is large, grouping firms into quartiles based on size reduces the level of detail and may mask important firm-level variation. This aggregation process can simplify the analysis but may also lead to a loss of information and potentially biased comparisons between groups.

Second, the data exhibits significant skewness and extreme values, particularly among larger firms. The presence of outliers may distort both visualizations and regression estimates, leading to results that are sensitive to extreme observations.

Third, the regression model may suffer from omitted variable bias. Only leverage and profit margin are included as explanatory variables, while other important factors such as firm age, industry differences, and macroeconomic conditions are not considered. This may limit the explanatory power and accuracy of the model.

Fourth, the assumptions of the OLS model are likely violated. Diagnostic results indicate non-normality in the residuals, suggesting that the estimated coefficients may not be fully reliable. This raises concerns about the robustness of statistical inference.

Finally, the analysis is based on cross-sectional data at a specific point in time. As a result, it does not capture dynamic changes in firm performance or causal relationships over time.

Overall, these limitations suggest that the findings should be interpreted with caution, and further analysis using more detailed data and advanced methods is recommended.
