## 1. Introduction

This project analyses the relationship between firm size, financial leverage, and profitability using financial data from Chinese listed companies.

Understanding how firm characteristics affect financial performance is important for investors when making informed decisions. This study focuses on key indicators such as Return on Assets (ROA), Return on Equity (ROE), and profit margin to evaluate firm performance from different perspectives.

## 2. Data Description

The dataset is obtained from the CSMAR database and contains financial information on Chinese listed companies.

For analytical purposes, firms are grouped into four size categories: Low, Medium-Low, Medium-High, and High. This grouping simplifies comparison across firms of different scales.

The key variables used in this analysis include:

- Net Profit: Measures the total earnings of a firm
- ROA (Return on Assets): Indicates how efficiently a firm uses its assets to generate profit
- ROE (Return on Equity): Reflects the return generated for shareholders
- Leverage: Represents the level of financial risk based on debt usage
- Profit Margin: Measures operational efficiency and profitability

Although the original dataset is large, aggregating firms into size groups reduces the number of observations, which may affect the robustness of the analysis.

```python
# Inspect the column structure of the CSMAR financial table 
financial_desc = db.describe_table(library="csmar", table="wrds_csmar_financial_master")
financial_desc
```

```python
# Select only the 3 most useful columns
financial_desc_small = financial_desc[["name", "type", "comment"]]

# Show the first 20 rows as a preview.
financial_desc_small.head(20)
```

```python
# Define the list of variable IDs we plan to use.
ids_today = [
    "a001000000",   # Total Assets
    "a002000000",   # Total Liabilities
    "a003000000",   # Total Shareholders' Equity
    "b001100000",   # Total Operating Revenue
    "b001200000",   # Total Operating Costs
    "b002000000"    # Net Profit
]
financial_desc_selected = financial_desc_small[financial_desc_small["name"].isin(ids_today)]

# Display the filtered description table.
financial_desc_selected
```

```python
# Store columns as a list so we can loop over them or join them into a SQL string.
financial_columns_list = ["stkcd", "accper", "typrep","a001000000","a002000000","a003000000", "b001100000", "b001200000","b002000000"]
```

```python
# Converts the format before we put it into the query.
financial_columns = ", ".join(financial_columns_list)

# Print the result to verify it looks like a valid SQL column list.
print(financial_columns)
print(type(financial_columns))
```

```python
# --- Build the SQL query for Chinese financial data using an f-string ---
financial_query = f"""
SELECT {financial_columns}
FROM csmar.wrds_csmar_financial_master
WHERE accper >= '2020-01-01'
AND typrep = 'A'
"""
print(financial_query)
```

```python
china_financial = db.raw_sql(financial_query, date_cols=["accper"])
china_financial
```

```python
# Add column Tto calculate size, ROA, ROE, leverage and profit margin
# ROA: profitability
# ROE: shareholder return
# Leverage: financial risk
china_financial['size'] = np.log(china_financial["a001000000"])
china_financial["ROA"] = china_financial["b002000000"] / china_financial["a001000000"]
china_financial["ROE"] = china_financial["b002000000"] / china_financial["a003000000"]
china_financial['leverage'] = china_financial["a002000000"] /china_financial["a001000000"]
china_financial['profit margin'] = china_financial["b002000000"] /china_financial["b001100000"]
china_financial
```

```python
china_financial["report_year"]  = china_financial["accper"].dt.year
china_financial["report_month"] = china_financial["accper"].dt.month
china_financial["report_day"]   = china_financial["accper"].dt.day
china_financial
```

```python
# Keep only year-end (December) reports
china_financial_annual = china_financial[china_financial['report_month'] == 12].copy()

china_financial_annual
```

```python
# The result shows us exactly what each of our chosen columns means.

selected_desc = financial_desc_small[financial_desc_small["name"].isin(financial_columns_list)]

# Display the filtered description — this is the reference for building rename_map.
selected_desc
```

```python
rename_map = dict(zip(selected_desc["name"], selected_desc["comment"]))

# Print the dictionary to verify the key → value pairs are correct.
rename_map
```

```python
china_financial_renamed = china_financial_annual.rename(columns=rename_map)

# Display the result — column names should now be readable English descriptions.
china_financial_renamed
```

```python
[value for value in rename_map.values()]
```

```python
selected_view = china_financial_renamed[[value for value in rename_map.values()]].copy()
selected_view["ROA"] = china_financial_renamed["ROA"]
selected_view["ROE"] = china_financial_renamed["ROE"]
selected_view["size"] = china_financial_renamed["size"]
selected_view["leverage"] = china_financial_renamed["leverage"]
selected_view["profit margin"] = china_financial_renamed["profit margin"]

selected_view
```

```python
# --- Check how many missing values (NaN) are in each column ---
selected_view.isna().sum()
```

```python
# Clean the data: Remove rows with missing values
selected_view = selected_view.dropna()
selected_view
```

```python
# Check again
selected_view.isna().sum()
```

```python
# Filter out companies with invalid total assets (assets <= 0)
selected_view = selected_view[selected_view['Total Assets'] > 0]
selected_view
```

```python
# Group companies by quartiles of company size
selected_view['size_group'] = pd.qcut(selected_view['size'], 4, labels=['Low', 'Medium-Low', 'Medium-High', 'High'])

selected_view[['size', 'size_group']]
```

```python
# Calculate the mean by company size group
average_values_by_size = selected_view.groupby('size_group',observed = True).agg({
    'Total Assets': 'mean',
    'Total Liabilities': 'mean',
    'Net Profit': 'mean',
    'ROA': 'mean',
    'ROE': 'mean',
    'leverage': 'mean',
    'profit margin':'mean'
}).reset_index()

average_values_by_size
```

```python
ai_table = average_values_by_size

# Display the table.
ai_table
```

```python
#  Select the column needed to be showed in the graph
select_column = average_values_by_size[['size_group','Net Profit','ROA', 'ROE', 'leverage','profit margin']]
select_column
```

```python
ai_table = select_column

# Display the table.
ai_table
```

# 5. Regression Analysis

To further examine the relationship between financial performance and risk, an Ordinary Least Squares (OLS) regression model is applied.

ROA is used as the dependent variable to measure profitability, while leverage and profit margin are included as independent variables.

The regression model aims to evaluate how financial risk and operational efficiency influence firm profitability.

The results suggest a positive relationship between leverage and ROA. However, the coefficients are not statistically significant, which is likely due to the small number of observations after data aggregation.

Therefore, the regression results should be interpreted with caution.

```python
import pandas as pd
import statsmodels.api as sm
df = ai_table.copy()

df = df[['ROA', 'leverage', 'profit margin']]


for col in df.columns:
    df[col] = df[col].astype(str).str.strip()   
    df[col] = pd.to_numeric(df[col], errors='coerce')


df = df.dropna()


print(df.dtypes)


Y = df['ROA']
X = df[['leverage', 'profit margin']]

X = sm.add_constant(X)

model = sm.OLS(Y, X).fit()
print(model.summary())
```

## The OLS regression model description
The model shows a high explanatory power, with an R-squared value of 0.919, indicating that approximately 91.9% of the variation in ROA can be explained by the independent variables.

The coefficient for leverage is positive (0.0727), suggesting that firms with higher leverage tend to achieve higher returns on assets. This may indicate that larger firms effectively utilize debt financing to enhance profitability.

In contrast, profit margin shows a negligible effect on ROA, with a coefficient close to zero, implying limited explanatory power in this model.

However, neither variable is statistically significant at conventional levels, as indicated by high p-values (greater than 0.05). This lack of significance is likely due to the extremely small sample size (only four observations), which limits the reliability and generalizability of the regression results.

Therefore, the findings should be interpreted with caution, and future analysis with a larger dataset is recommended to obtain more robust conclusions.
